## 数据的存储与读取

通用的数据格式，可以在不同语言中加载和存储

本节简单了解两种数据存储结构 csv 和 json

1、csv 格式

由逗号将数据分开的字符序列，可以由 excel 打开

- 读取

```python
with open('example.csv', 'r', encoding='utf-8') as f:
    ls = []
    for line in f:                                # 逐行读取
        ls.append(line.strip('\n').split(','))    # 去掉每行的换行符，然后用','进行分割
for res in ls:
    print(res)
```

- 写入

```python
ls = [['id', 'category', 'query1', 'query2', 'label'],['0', '咳血', '"剧烈运动后咯血', '是怎么了?"', '剧烈运动后咯血是什么原因？', '-1'],['1', '咳血', '"剧烈运动后咯血', '是怎么了?"', '剧烈运动后为什么会咯血？', '-1'],['2', '咳血', '"剧烈运动后咯血', '是怎么了?"', '剧烈运动后咯血，应该怎么处理？', '-1']]
with open('example.csv', 'w', encoding='utf-8') as f:
    for row in ls:                   # 逐行写入
        f.write(','.join(row)+'\n')  # 用逗号组合成字符串形式，末尾加换行符
```

也可以借助 csv 模块完成上诉操作

2、json 格式

常用来存储字典类型

- 写入————dump()

```python
import json

scores = {'Petter':{'math':96, 'physics':98},
         'Paul':{'math':92, 'physics':99},
         'Mary':{'math':98, 'physics':97}}
with open('score.json', 'w', encoding='utf-8') as f:
    # indent 表示字符串换行+缩进 ensure_ascii=False 显示中文
    json.dump(scores, f, indent=4, ensure_ascii=False)
```

- 读取————load()

```python
with open('score.json', 'r', encoding='utf-8') as f:
    scores = json.load(f)
    for k,v in scores.items():
        print(k,v)
```

### 异常的处理

提升程序的稳定性和可靠性

1、try_except

- 如果 try 内代码块顺利执行，except 不被触发
- 如果 try 内代码块发生错误，触发 except 执行 except 内代码块

```python
# 单分支

x = 10
y = 0
try:
    z = x/y
except ZeroDivisionError:          # 一般来说会预判到出现什么错误
    # z = x/(y+1e-7)
    # print(z)
    print('0不可以被除！！！')
```

```python
# 多分支

ls = []
d = {'name': 'Jack'}
try:
    # y = x
    # ls[3]
    d['age']
except NameError:
    print('变量名不存在！！！')
except IndexError:
    print('索引超出界限！！！')
except KeyError:
    print('键不存在！！！')
```

- 万能异常 Exception （所有错误的老祖宗）

```python
ls = []
d = {'name': 'Jack'}
try:
    # y = x
    # ls[3]
    d['age']
except Exception:
    print('出错了！！！')
```

- 捕获异常的值 as

```python
ls = []
d = {'name': 'Jack'}
try:
    y = x
    ls[3]
    d['age']
except Exception as e:     # 虽然不能获得错误具体类型，但可以获得错误的值
    print(e)
```




2、try_except_else

- 如果 try 模块执行，则 else 模块也执行

可以将 else 看做 try 成功的额外奖赏

```python
try:
    with open('浪淘沙_北戴河.txt') as f:
        text = f.read()
except FileNotFoundError:
    print('找不到该文件，美颜了？')
else:
    for s in ['\n', '，', '。', '？']:     # 去掉换行符和标点符号
        text = text.replace(s, '')
    print('毛主席的名作《浪淘沙·北戴河》共有 {} 个字。'.format(len(text)))
```

3、try_except_finally

- 不论 try 模块是否执行，finally 最后都执行

```python
ls = []
d = {'name': 'Jack'}
try:
    y = x
    ls[3]
    d['age']
except Exception as e:
    print(e)
finally:
    print('不论是否触发异常，都会执行。')
```

## 题目一（文件的读写）：

1、读取txt文件，统计文本中所有字符的出现频次，将结果按如下格式保存到文件“xxx_字符统计.txt”

- 令：3904，狐：3904，冲：4728

```python
def count_word(s):  
    """统计字符频次"""
    d = {}
    for i in s:
        d[i] = d.get(i, 0) + 1
    return d


with open('E:\资料\网络文学\完美世界.txt', 'r', encoding='gbk') as f:
    s = f.read()
    dic = count_word(s)
    
with open('完美世界_字符统计.txt', 'w', encoding='gbk') as f:
    for k, v in dic.items():
        f.writelines('{}: {}\n'.format(k, v))
```
